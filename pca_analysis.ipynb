# Principal Component Analysis Dimension Reduction Techinique

```python
#importing basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the and viewing the data
df=pd.read_csv('D:\Data analytics project\hyperparameter tuning\data.csv')
df.head()
```

```python
#Dividing data into dependent and target varibles
X=df.drop('diagnosis',axis=1)
y=df.diagnosis
```

```python
#reducing X columns to make the demonstration of PCA more effective
X=X[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']]
```

```python
X.head()
```

```python
#doing feature scaling to impliment pca effectively
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled_X=scaler.fit_transform(X)
```

```python
#looking at the scaled features
scaled_X=pd.DataFrame(scaled_X,columns=['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean'])
scaled_X
```

```python
#importing pca and fitting to our features
from sklearn.decomposition import PCA
#the value in the PCA demonstrate the reductoin dimension 
pca=PCA(2)
x_pca=pca.fit_transform(scaled_X)
```

```python
#hence the finally we got the resulted array of fimal dimension
x_pca.shape
```

```python
#plotting 2 the dimension of the result
plt.figure(figsize=(12,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=y);
plt.xlabel('first principal component')
plt.ylabel('second principal component')
```

As we can see in the graph we can see how the pca algorithm has work to select the feature
